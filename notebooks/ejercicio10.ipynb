```python
import sys
import os
sys.path.append(os.path.abspath("../src"))
from src.funciones import *

rounds = [
{
'Shadow': {'kills': 2, 'assists': 1, 'deaths': True},
'Blaze': {'kills': 1, 'assists': 0, 'deaths': False},
'Viper': {'kills': 1, 'assists': 2, 'deaths': True},
'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 0, 'assists': 2, 'deaths': False},
'Blaze': {'kills': 2, 'assists': 0, 'deaths': True},
'Viper': {'kills': 1, 'assists': 1, 'deaths': False},
'Frost': {'kills': 2, 'assists': 1, 'deaths': True},
'Reaper': {'kills': 0, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 1, 'assists': 0, 'deaths': False},
'Blaze': {'kills': 2, 'assists': 2, 'deaths': True},
'Viper': {'kills': 1, 'assists': 1, 'deaths': True},
'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 2, 'assists': 1, 'deaths': False},
'Blaze': {'kills': 1, 'assists': 0, 'deaths': True},
'Viper': {'kills': 0, 'assists': 2, 'deaths': False},
'Frost': {'kills': 1, 'assists': 1, 'deaths': True},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 1, 'assists': 2, 'deaths': True},
'Blaze': {'kills': 0, 'assists': 1, 'deaths': False},
'Viper': {'kills': 2, 'assists': 0, 'deaths': True},
'Frost': {'kills': 1, 'assists': 1, 'deaths': False},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': True}
}
]

# Diccionario para almacenar las estadisticas de los jugadores
players_data = {}
i = 1
for round in rounds:
    # Puntos para cada jugador en la ronda
    mvp_player = None
    best_points = -100
    # Cargamos un nuevo jugador a la partida
    for player in round:
        # Guardo los puntos totales del jugador actual en la ronda actual
        point = score(round[player]['kills'], round[player]['assists'], round[player]['deaths'])

        # Si el jugador ya esta cargado le sumamos las estadisticas
        if player in players_data:
            update_stats(round, player, players_data, point)

        # Caso que el jugador aun no este cargado lo agrego al diccionario
        else:
            new_player(round, player, players_data, point)

        # Busco el mvp de la ronda
        mvp_player , best_points = find_mvp(point, best_points, player, mvp_player)

    # Selecciono al mvp de la ronda    
    players_data[mvp_player]['MVPs'] += 1

    # Ordeno la ronda actual
    ranking = ordenar_ranking(players_data)

    # se imprime el ranking
    print_ranking(ranking, i)
    i += 1
print_ranking(ranking, i)
```
